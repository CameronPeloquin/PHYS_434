```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats


bkgd = stats.norm.rvs(scale=3, size=1000000)
signal = np.random.uniform(size=1000000, low=0, high=20)

data = bkgd + signal

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,signal, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
```

```python
# Problem 1.B
# Choose injected signal of 10.2564
print(temp2[20])
slice = Psd[:,20]
plt.stairs(slice, temp)
plt.title('P(d|s) for a true signal of 10.2564', fontsize=24)
plt.xlabel('Value', fontsize=24)
plt.yscale('log')
plt.show()
```

Problem 1.B

This plot is a slice of the 2d Histogram above, it shows the probability densities of all the data values where the injected signal is equal to 10.2564. This means that it if we get a true signal value of 10.2564, this graph shows that probabilities that an observed data value would result in that signal.

```python
# Problem 1.C
# Choose observed data value of 9.7462
print(temp[33])
slice = Psd[33,:]
plt.stairs(slice, temp2)
plt.title('P(s|d) for Observed Data Value of 9.7462', fontsize=24)
plt.xlabel('Value', fontsize=24)
plt.yscale('log')
plt.show()
```
