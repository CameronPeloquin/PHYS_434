```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import math
```

```python
# Problem 0:

t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

```python
# Problem 1.A:

first_1000_bkgd = bkgd[0:1000]
first_1000_t = t[0:1000]

plt.scatter(first_1000_t, first_1000_bkgd)
```

```python
# Problem 1.B:
bkgd_min = 10000 
bkgd_max = 20000

first_1000_bkgd = bkgd[bkgd_min:bkgd_max]
first_1000_t = t[bkgd_min:bkgd_max]

plt.scatter(first_1000_t, first_1000_bkgd)
```

```python
# Problem 1.C:

every_100_bkgd = bkgd[0:100000:100]
every_100_t = t[0:100000:100]

plt.scatter(every_100_t, every_100_bkgd)
```

```python
# Problem 2.A:

plt.hist2d(t,bkgd,bins=10,density=True)
plt.colorbar()
```

Problem 2.B

This plot shows the density of points. This means that the colors represent the number of points within a bin divided by the total number of data points. This allows someone to clearly see where most of the points lie, as the bins that have more overlapping points are shown as brighter colors, than bins that contain less points.
