How Data was Collected

The data for this experiment was collected at the Large Hadron Collider at CERN. The collider is made up of a 27 kilometer ring made of superconducting magnets. Within the ring are two vaccuum tubes that send groups of particles in opposite directions around the ring. There are four detectors around the ring, ALICE, ATLAS, CMS, and LHCb. Inside of these detectors, the two vaccuum tubes converge into one, smaller tube, to increase the chances of protons collidig. At the point of collision the particles are traveling at %99 the speed of light. The LHC produced around 1 billion colllisions per second, which is far too much data to store and process, so to cut down on the amount of data received, triggers are used to filter out the vast majority of data, allowing only around 100 data points to be taken per billion events detected. Particle collisons cause jets to form. The most interesting collisions to look at are hard collisions between heavier particles, as this increases the chance of producing rare particles. These rare particles are generally unstable, and decay into lighter particles quickly, and often decay before reaching the detector. This is where detecting events such as a higgs event gets difficult. The Higgs for example will decay into (What the higgs decays into) before reaching the detector. This means that the detector needs to distinguish between lighter particles that were from a higgs decaying, compared to them forming from other collisions. This is where the 14 variables in the dataset that describe jets come in. These help distinguish the signal from the backgound.


What are Jets:

Jets are cones of particles that are ejected from a collision, the particles in the jet are produced by the hadronization of a quark or gluon, which is the process where hadrons are created from quarks and gluons. When an object containing color charge fragments, each fragment carries away some of the color charge. In order to obey confinement, these fragments create other colored objects around them to form colorless objects. All of these objects tend to travel in close to the same direction, hence the narrow cone. So these objects that form are what make up jets.

The fifth and sixth varibles are ee2, and ee3 denoted as $e_{2}$ and $e_{3}$. These values are Energy Correlation Ratios.

The equation below is the energy correlation

$E_{CFN}(\beta)$ = $\sum_{i<j \in J}$ $p_{Ti}p_{Tj}(\Delta$$R_{ij})^\beta$

$e_{2}$ and $e_{3}$ are the ratios of the energy correlations:

$e_{2}$ = $\frac{E_{CF2}}{E_{CF1}}$

$e_{3}$ = $\frac{E_{CF3}}{E_{CF2}}$

The above plots show  histograms of the ee2 and ee3 values for both the higgs signal dataset, and the QCD background dataset.

The above five variables are the subjettiness variables. Subjettiness is a measure of how a jet can be made up of subjets

$\tau_{N}$ = $\frac{1}{\tau_{0}}$$\sum_{i}\in Jp_{Ti}$min($\Delta$$R_{1}...R{N}$)

N is the subjet axis and can take on integer values, and $R_{N}$ is the distance between the $i$th subjet, and the subjet axis N.

$\tau_{21}$, and $\tau_{32}$ are the ratios of $\tau_{N}$ values between consecutive values of N

$\tau_{21}$ = $\frac{\tau_{2}}{\tau_{1}}$

$\tau_{32}$ = $\frac{\tau_{3}}{\tau_{2}}$

Like $D_{2}$ these ratios are useful in determining two body substructures in jets.

The above plots show histograms of the t1, t2, t3, t21, and t32 values for both the higgs signal dataset, and the QCD background dataset.

Correlation:

Because many of these variables are built off of eachother, it makes sense that some would be correlated. The plots below show the correlation between mass and pT, mass and ee2, mass and ee3, d2 with ee2 and ee3, and t21, and t32. It also makes sense that there isn't correlation between phi and any other variable as it is not uniformly distributed.

```python
def makeHists(var1, var2):
    fig, ax = plt.subplots(1,2, figsize = (20,10))
    ax[0].hist2d(qcd_dict[var1], qcd_dict[var2], bins=[50,50])
    ax[1].hist2d(higgs_dict[var1], higgs_dict[var2], bins=[50,50])
    ax[0].set_title('QCD')
    ax[1].set_title('Higgs')
    ax[0].set_xlabel(var1)
    ax[0].set_ylabel(var2)
    ax[1].set_xlabel(var1)
    ax[1].set_ylabel(var2)


makeHists('mass', 'ee2')
makeHists('mass', 'ee3')
makeHists('mass', 'pt')
makeHists('d2', 'ee3')
makeHists('d2', 'ee2')
makeHists('t21', 't32')
makeHists('phi', 'eta')
```

Calculating discovery significance:

In order to be sure that a higgs event has occured, we need to insure that the number of higgs events occurring in a dataset it above the five sigma threshold for higgs events occurring in the background. Due to this being a particle physics experiment, the number of higgs events occurring will follow a poisson distribution, and can be represented by a poisson distribution that has a mean equal to the expected yield of higgs events.

```python
sigma_5_prob = 3.5*10**-6
expected_yield_qcd = 2000
threshold = stats.poisson.isf(q=sigma_5_prob,mu=expected_yield_qcd)
print(threshold)
```

Without any data cuts, we would need to observe 2204 higgs events in the QCD background to ensure the discovery of a higgs boson.
